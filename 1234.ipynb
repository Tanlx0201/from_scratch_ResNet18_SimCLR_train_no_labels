```python
from torchvision import datasets, transforms
import torch

transform = transforms.Compose([
    transforms.ToTensor()
])

dataset = datasets.STL10(
    root="./data",
    split="train",
    download=True,
    transform=transform
)

# Lấy 1 ảnh
image, label = dataset[0]
# image shape: (3, 96, 96)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Flatten pixel
x = image.view(-1).numpy()
y = np.maximum(0, x)

plt.figure()
plt.hist(x, bins=100, alpha=0.5, label="Before ReLU")
plt.hist(y, bins=100, alpha=0.5, label="After ReLU")
plt.legend()
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.title("ReLU applied on STL-10 image pixels")
plt.show()
```

```python
import re

log = open("train_resnet.txt").read()  # hoặc copy log vào chuỗi

epochs, train_acc, val_acc = [], [], []

for e, ta, va in re.findall(
    r"Epoch\s+(\d+).*?Train Acc\s+([\d.]+).*?Val Acc\s+([\d.]+)", log
):
    epochs.append(int(e))
    train_acc.append(float(ta))
    val_acc.append(float(va))

plt.figure()
plt.plot(epochs, train_acc, label="Training Accuracy")
plt.plot(epochs, val_acc, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Training & Validation Accuracy on STL-10")
plt.legend()
plt.grid(True)
plt.show()
```
